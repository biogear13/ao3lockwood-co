```python
def scrape_website(url):
    try:
        browser.get(url)
        time.sleep(5)
        # Scroll down repeatedly until reaching the end of the webpage
        while True:
            # Scroll to the bottom of the page
            browser.find_element_by_tag_name('body').send_keys(Keys.END)
            time.sleep(3)  # Wait for the page to load
            
            # Check if reached the end of the page
            scroll_height = browser.execute_script("return document.body.scrollHeight")
            client_height = browser.execute_script("return document.documentElement.clientHeight")
            if scroll_height == client_height:
                break
        # Get the page source
        return browser.page_source
    except:
        print("Error in scraping the website")
        return None

def get_title(page_source):
    try:
        soup = BeautifulSoup(page_source, 'html.parser')
        # get hrefs
        links = soup.find_all("a", href=lambda href: href and href.startswith("/title/"))
        # get titles
        titles = [link.text for link in links]
        # remove '/title/' from titles
        return titles
    except:
        print("Error in getting the title")
        return None

def get_unique_titles(final_list, new_list):
    try:
        for title in new_list:
            if title not in final_list:
                final_list.append(title)
        return final_list
    except:
        print("Error in getting the unique titles")
        return None
```

```python
final_list = []
for key, value in links.items():
    browser.get(value)
    time.sleep(5)
    page_source = browser.page_source
    new_list = get_title(page_source)
    final_list = get_unique_titles(final_list, new_list)
    print(f"Scraped {key} successfully")
    
browser.close()
print(final_list)
```
