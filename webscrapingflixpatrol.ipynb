```python
def get_shows(link, df):
    source = requests.get(link, headers={
                              'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64)'}).text
    soup = BeautifulSoup(source,'html.parser')
    index = len(df)
    shows = soup.find_all('tr', attrs={'class':'table-group'})
    for show in shows:
        print(index)
        df.loc[index, 'release_date'] = show.find('td', attrs={'class':"table-td w-24 text-left whitespace-nowrap"}).text
        df.loc[index, 'title'] = show.find('div', attrs={'class':"group-hover:underline"}).text
        try:
            description = show.find('div', attrs={'class':"flex flex-wrap text-sm leading-6 text-gray-500"}).text.split('|')
            df.loc[index, 'location'] = description[1]
            df.loc[index, 's1release_date'] = description[2]
            df.loc[index,'genre'] = description[4]
        except:
            df.loc[index, 'location'] = np.nan
            df.loc[index, 's1release_date'] = np.nan
            df.loc[index,'genre'] = np.nan
        try:
            df.loc[index, 'imdb'] = show.find('td', attrs={'class':"table-td hidden md:table-cell text-gray-500 table-hover:text-gray-400 whitespace-nowrap"}).text
            #<td class="table-td hidden md:table-cell text-gray-500 table-hover:text-gray-400 whitespace-nowrap"> <div class="flex items-center"> <span class="inline-block align-baseline mr-2 h-4 w-4 platform-imdb"></span> 4.5/10 </div> </td>
        except:
            df.loc[index, 'imdb'] = np.nan
        try:
            df.loc[index, 'rottentomato'] = show.find_all('div', attrs={'class':"flex items-center"})[1].text
        except:
            df.loc['index', 'rottentomato'] = np.nan
        try:
            df.loc[index, 'ranking'] = show.find('td', attrs={'class':"table-td hidden sm:table-cell text-right text-gray-400 font-semibold tabular-nums whitespace-nowrap"}).text
        except:
            df.loc[index, 'ranking'] = np.nan
        #print(df.loc[index])
        index += 1
    time.sleep(10)    
    return df
```

```python
link='https://flixpatrol.com/calendar/new/tv-shows/netflix/'
df = pd.DataFrame(columns=['release_date', 'title', 'location', 's1release_date', 'genre', 'imdb', 'rottentomato', 'ranking'])
df = get_shows(link, df)
df
```

```python
for n in range(2,65):
    p_link = f'{link}/right-now/{n}/'
    df = get_shows(p_link, df)
    df.to_csv('netflixshows.csv', index=False)
    time.sleep(10)
df
```

```python
df['imdb'] = df['imdb'].str.replace('/10', '')
df['rottentomato'] = df['rottentomato'].str.replace('%', '')
df['ranking'] = df['ranking'].str.replace('#', '').replace(',', '')
df['title'] = df['title'].str.replace(')', '')
df['title'], df['season'] = df['title'].str.split('(', 1).str
df
```
